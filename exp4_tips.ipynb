```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load the Titanic dataset
file_path = "/content/exp4_tips.csv"
df = pd.read_csv(file_path)
```

```python
# Set a visual style
sns.set_style("whitegrid")
```

```python
# Histograms for numerical features
df.hist(figsize=(12, 8), bins=20, edgecolor='black', layout=(3, 3))
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.show()
```

```python
# Box plots for Total Bill to identify outliers
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.boxplot(y=df["total_bill"], ax=axes[0], palette="pastel") # Changed column to 'total_bill'
axes[0].set_title("Box Plot of Total Bill") # Changed title

sns.boxplot(y=df["tip"], ax=axes[1], palette="pastel") # Changed column to 'tip'
axes[1].set_title("Box Plot of Tip") # Changed title
plt.show()
```

```python
# Heatmap to show correlations
plt.figure(figsize=(10, 6))
# Calculate correlations for numerical columns only
numerical_df = df.select_dtypes(include=['number'])  # Select only numerical columns
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Bar plot for total_bill by sex (assuming 'sex' column exists)
plt.figure(figsize=(8, 5))
sns.barplot(x="sex", y="total_bill", data=df, palette="coolwarm")  # Changed columns to 'sex' and 'total_bill'
plt.title("Total Bill by Sex")  # Changed title
plt.ylabel("Total Bill")  # Changed y-axis label
plt.show()
```

```python
# Bar plot for total_bill by sex (assuming 'sex' column exists)
plt.figure(figsize=(8, 5))
# Assuming 'day' and 'time' are columns in your DataFrame
sns.barplot(x="day", y="tip", data=df, palette="coolwarm")
plt.title("Tip amount by Day")
plt.ylabel("Tip Amount")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
# Create a stacked bar plot with 'sex' on x-axis and 'smoker' as the hue
sns.histplot(data=df, x="sex", hue="smoker", multiple="stack", palette="coolwarm", discrete=True)

plt.title("Smoking Distribution by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
# Generate a word cloud for a different column, e.g., 'day'
# If you have a text column you want to analyze, replace 'day' with that column name
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(" ".join(df["day"].astype(str)))
# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Days") # Changed title to reflect the change in column
plt.show()
```
